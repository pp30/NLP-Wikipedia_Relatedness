#WordSimilarity

```
p = 'Evaluation_Benchmarks/Word_Similarity'
p1 = p+ '/WS353_Exact/EN-WS-353-ALL.txt'
p2 = p + '/RG65/EN-RG-65.txt'
p3 = p + '/MEN/EN-MEN-TR-3k.txt'

p1_w = 'ws353_word_string.txt'
p2_w = 'rg_word_string.txt'
p3_w = 'men_word_string.txt'

WS = open(p1,'r')
RG = open(p2,'r')
MEN = open(p3,'r')

count=0
word = ''
print(len(word))
for index,line in tqdm(enumerate(WS)):
  #if count ==2: break
  count+=1
  line = line.strip().lower().split()
  word = word + line[0] + ' ' + line[1] + ' '

file = open(p1_w,'w')
file.write(word)
file.close()

count=0
word2 = ''
#print(len(word))
for index,line in tqdm(enumerate(RG)):
  #if count ==2: break
  count+=1
  line = line.strip().lower().split()
  word2 = word2 + line[0] + ' ' + line[1] + ' '

file = open(p2_w,'w')
file.write(word2)
file.close()

count=0
word3 = ''
#print(len(word))
for index,line in tqdm(enumerate(MEN)):
  #if count ==2: break
  count+=1
  line = line.strip().lower().split()
  word3 = word3 + line[0] + ' ' + line[1] + ' '

file = open(p3_w,'w')
file.write(word3)
file.close()
```

#SA

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```
psa_tr = 'Evaluation_Benchmarks/Sentiment_Analysis/Dataset_Exact/train.txt'
psa_ts = 'Evaluation_Benchmarks/Sentiment_Analysis/Dataset_Exact/test.txt'

psa_tr_w = 'SA-train_word_string.txt'
psa_ts_w = 'SA-test_word_string.txt'

tr_obj = open(psa_tr,'r')
ts_obj = open(psa_ts,'r')
count = 0
str_word = ''
for index,line in tqdm(enumerate(tr_obj)):
  #if count ==20: break
  count += 1
  line = line.strip().split()
  #temp = []
  for word in line[1:]:
    #print("word:", word)
    if word.isalpha() and word not in stop_words:
      #print("hahahaha")
      str_word = str_word + word.lower() + ' '

tr_obj.close()

file = open(psa_tr_w,'w')
file.write(str_word)
file.close()

count = 0
str_word2 = ''
for index,line in tqdm(enumerate(ts_obj)):
  #if count ==20: break
  count += 1
  line = line.strip().split()
  #temp = []
  for word in line[1:]:
    #print("word:", word)
    if word.isalpha() and word not in stop_words:
      #print("hahahaha")
      str_word2 = str_word2 + word.lower() + ' '

ts_obj.close()

file = open(psa_ts_w,'w')
file.write(str_word2)
file.close()



```

```
total_sa_string = str_word+str_word2
```
