```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from gensim.models import KeyedVectors
from tqdm.notebook import tqdm
import math
```

```
path_1 = 'Evaluation_Benchmarks/Sentiment_Analysis/Dataset_Exact/train.txt' # path to Stanford Treebank Sentiment analysis (SA) train dataset file
path_2 = '/'.join(path_1.split('/')[:-1])+ '/test.txt'  # path to Stanford Treebank Sentiment analysis (SA) test dataset file
path_3 = '/'.join(path_1.split('/')[:-1])+ '/dev.txt' # path to Stanford Treebank Sentiment analysis (SA) developement dataset file 

obj1 = open(path_1,'r')
obj2 = open(path_2,'r')
obj3 = open(path_3,'r')
 
X_train_sentence = [] 
Y_train = []

X_test_sentence = []
Y_test = []

X_dev_sentence = []
Y_dev = []

for index, line in enumerate(obj2):
  line = line.strip().split()
  Y_test.append(int(line[0]))
  X_test_sentence.append([word.lower() for word in line[1:]])

for index, line in enumerate(obj1):
  line = line.strip().split()
  Y_train.append(int(line[0]))
  X_train_sentence.append([word.lower() for word in line[1:]])

for index, line in enumerate(obj3):
  line = line.strip().split()
  Y_dev.append(int(line[0]))
  X_dev_sentence.append([word.lower() for word in line[1:]])
print('Loaded all text files into memory')

def train_logistic_model(word_vec, vector_dimension):
  """
  Computes test accuracy for the task of labelling a movie review sentence (Postive/Negative) using Logistic Regression

  Parameters
  ----------
  word_vec : dict
    A dictionary with keys as words (dtype = string) and values as LIST of floating point numbers from given word vector file

  vector_dimensions : int 
    The length of the vectors present in given word-vector file  

  Returns
  -------
  arg1 : float
    train accuracy score (set to 0 by default as currently not needed for the task at hand)

  arg2 : float
    test accuracy score 
  
  arg3 : float
    Developement accuracy score (set to 0 by default as currently not needed for the task at hand)  

  """

  X_train = np.zeros([len(X_train_sentence),vector_dimension])
  X_test = np.zeros([len(X_test_sentence),vector_dimension])
  X_dev = np.zeros([len(X_dev_sentence),vector_dimension])

  print('Developing train data')
  for index, sentence in tqdm(enumerate(X_train_sentence)):
    sentence_vector = np.zeros([vector_dimension])
    count = 0
    for word in sentence:
      try: 
        sentence_vector = sentence_vector + word_vec[word]
        count+=1
      except: 
        temp=0
        count+=1 # Comment this line out and try running this cell.
    X_train[index,:] = sentence_vector/count

  print('Developing test data')
  for index, sentence in tqdm(enumerate(X_test_sentence)):
    sentence_vector = np.zeros([vector_dimension])
    count = 0
    for word in sentence:
      try: 
        sentence_vector = sentence_vector + word_vec[word]
        count+=1
      except:
        temp=0
        count+=1 # Comment this line out and try running this cell.
    X_test[index,:] = sentence_vector/count

  print('Developing dev data')
  for index, sentence in tqdm(enumerate(X_dev_sentence)):
    sentence_vector = np.zeros([vector_dimension])
    count = 0
    for word in sentence:
      try: 
        sentence_vector = sentence_vector + word_vec[word]
        count+=1
      except: 
        temp=0
        count+=1 # Comment this line out and try running this cell.
    X_dev[index,:] = sentence_vector/count


  # model fitting code

  model = LogisticRegression(random_state=0)
  model.fit(X_train, Y_train)
  accuracy_score_test = model.score(X_test, Y_test)
  print('Parameters :', np.sum(model.decision_function(X_test)>0))
  # print(Y_train)
  print(np.shape(X_train), np.shape(X_test), np.shape(X_dev))
  accuracy_score_train=0 
  # accuracy_score_test=0 
  accuracy_score_dev=0

  return accuracy_score_train, accuracy_score_test, accuracy_score_dev

```

```
#Global Context word vectors file
WordVec_path_a = 'Vector Representation of Words/3/3_word_vectors.txt'
#Skip Gram (SG) word vectors file
WordVec_path_b = 'Vector Representation of Words/2/GoogleNews-vectors-negative300.bin'
#Glove word vector file
WordVec_path_c = 'Vector Representation of Words/1/glove.6B.300d.txt'

#Global Context retrofitted word vectors file
outfile_path_a1 = 'Vector Representation of Words/3/RETRO_FrameNet_3_word_vectors.txt' 
outfile_path_a2 = 'Vector Representation of Words/3/RETRO_ppdb-xl_3_word_vectors.txt' 
outfile_path_a3 = 'Vector Representation of Words/3/RETRO_wordnet-synonyms_3_word_vectors.txt' 
outfile_path_a4 = 'Vector Representation of Words/3/RETRO_wordnet-synonyms+_3_word_vectors.txt' 

#Glove retrofitted word vector files
outfile_path_c1 = 'Vector Representation of Words/1/RETRO_FrameNet_glove.6B.300d.txt'
outfile_path_c2 = 'Vector Representation of Words/1/RETRO_ppdb-xl_glove.6B.300d.txt'
outfile_path_c3 = 'Vector Representation of Words/1/RETRO_wordnet-synonyms_glove.6B.300d.txt'
outfile_path_c4 = 'Vector Representation of Words/1/RETRO_wordnet-synonyms+_glove.6B.300d.txt'
```

```
'''Evaluating on the Glove and GC word vectors file (including retrofitted Glove and GC word vectors)'''

''' Read all the word vectors and normalize them '''
def read_word_vecs(filename):
  # Use this for Glove and Global Context Vectors
  wordVectors = {}
  fileObject = open(filename, 'r')
  for index,line in tqdm(enumerate(fileObject)):
    # if index%50000==0:print('->', index,'/ 3000000')
    line = line.strip().lower()
    word = line.split()[0]
    wordVectors[word] = np.zeros(len(line.split())-1, dtype=float)
    for index, vecVal in enumerate(line.split()[1:]):
      wordVectors[word][index] = float(vecVal)
    ''' normalize weight vector '''
    wordVectors[word] /= math.sqrt((wordVectors[word]**2).sum() + 1e-6)
    
  return wordVectors

''' change 2nd argument inside tuple to 50 or 300 accordingly for Global context (50) and Glove word vector files respectively '''   
filenames = [(outfile_path_a4 ,50), (outfile_path_c4,300)]

for filename in filenames:
  word_vec = read_word_vecs(filename[0])
  print('word vec generated for :', filename)
  _,accuracy_score_test,_ = train_logistic_model(word_vec, vector_dimension=filename[1])
  print('Accuracy :',accuracy_score_test)
```

```
'''Evaluating on the SG (skip gram) word vectors file (does not include retrofitted SG word vectors)'''

filename = WordVec_path_b 
def read_word_vectors_skipgram():
  # Use this for Google skipgram word2vec model
  word_vec_bin = KeyedVectors.load_word2vec_format(filename, binary=True)
  return word_vec_bin

word_vec = read_word_vectors_skipgram()
print('word vec generated for :', filename)
_,accuracy_score_test,_ = train_logistic_model(word_vec, vector_dimension=300)
print('Accuracy :',accuracy_score_test)
```
